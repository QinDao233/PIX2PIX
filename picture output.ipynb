```python
import os
import sys
import time
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import functools
import torch.nn.init as init

# 设置随机种子
torch.manual_seed(42)
np.random.seed(42)

# -------------------------- 模型定义（保持不变） --------------------------
# （省略模型定义部分，与之前完全一致）
class GANLoss(nn.Module):
    """定义GAN损失函数"""
    def __init__(self, gan_mode, target_real_label=1.0, target_fake_label=0.0):
        super(GANLoss, self).__init__()
        self.register_buffer('real_label', torch.tensor(target_real_label))
        self.register_buffer('fake_label', torch.tensor(target_fake_label))
        self.gan_mode = gan_mode
        if gan_mode == 'lsgan':
            self.loss = nn.MSELoss()
        elif gan_mode == 'vanilla':
            self.loss = nn.BCEWithLogitsLoss()
        elif gan_mode in ['wgangp']:
            self.loss = None
        else:
            raise NotImplementedError('gan mode %s not implemented' % gan_mode)

    def get_target_tensor(self, prediction, target_is_real):
        if target_is_real:
            target_tensor = self.real_label
        else:
            target_tensor = self.fake_label
        return target_tensor.expand_as(prediction)

    def __call__(self, prediction, target_is_real):
        if self.gan_mode in ['lsgan', 'vanilla']:
            target_tensor = self.get_target_tensor(prediction, target_is_real)
            loss = self.loss(prediction, target_tensor)
        elif self.gan_mode == 'wgangp':
            if target_is_real:
                loss = -prediction.mean()
            else:
                loss = prediction.mean()
        return loss


class ResnetBlock(nn.Module):
    """Resnet块定义"""
    def __init__(self, dim, padding_type, norm_layer, use_dropout, use_bias):
        super(ResnetBlock, self).__init__()
        self.conv_block = self.build_conv_block(dim, padding_type, norm_layer, use_dropout, use_bias)

    def build_conv_block(self, dim, padding_type, norm_layer, use_dropout, use_bias):
        conv_block = []
        p = 0
        if padding_type == 'reflect':
            conv_block += [nn.ReflectionPad2d(1)]
        elif padding_type == 'replicate':
            conv_block += [nn.ReplicationPad2d(1)]
        elif padding_type == 'zero':
            p = 1
        conv_block += [
            nn.Conv2d(dim, dim, kernel_size=3, padding=p, bias=use_bias),
            norm_layer(dim),
            nn.ReLU(True)
        ]
        if use_dropout:
            conv_block += [nn.Dropout(0.5)]

        p = 0
        if padding_type == 'reflect':
            conv_block += [nn.ReflectionPad2d(1)]
        elif padding_type == 'replicate':
            conv_block += [nn.ReplicationPad2d(1)]
        elif padding_type == 'zero':
            p = 1
        conv_block += [
            nn.Conv2d(dim, dim, kernel_size=3, padding=p, bias=use_bias),
            norm_layer(dim)
        ]
        return nn.Sequential(*conv_block)

    def forward(self, x):
        return x + self.conv_block(x)


class ResnetGenerator(nn.Module):
    """Resnet生成器"""
    def __init__(self, input_nc, output_nc, ngf=64, norm_layer=nn.BatchNorm2d, 
                 use_dropout=False, n_blocks=6, padding_type='reflect'):
        assert(n_blocks >= 0)
        super(ResnetGenerator, self).__init__()
        if type(norm_layer) == functools.partial:
            use_bias = norm_layer.func == nn.InstanceNorm2d
        else:
            use_bias = norm_layer == nn.InstanceNorm2d

        model = [nn.ReflectionPad2d(3),
                 nn.Conv2d(input_nc, ngf, kernel_size=7, padding=0, bias=use_bias),
                 norm_layer(ngf),
                 nn.ReLU(True)]

        n_downsampling = 2
        for i in range(n_downsampling):
            mult = 2 ** i
            model += [nn.Conv2d(ngf * mult, ngf * mult * 2, kernel_size=3, stride=2, padding=1, bias=use_bias),
                      norm_layer(ngf * mult * 2),
                      nn.ReLU(True)]

        mult = 2 ** n_downsampling
        for i in range(n_blocks):
            model += [ResnetBlock(ngf * mult, padding_type=padding_type, norm_layer=norm_layer, 
                                 use_dropout=use_dropout, use_bias=use_bias)]

        for i in range(n_downsampling):
            mult = 2 ** (n_downsampling - i)
            model += [nn.ConvTranspose2d(ngf * mult, int(ngf * mult / 2),
                                         kernel_size=3, stride=2,
                                         padding=1, output_padding=1,
                                         bias=use_bias),
                      norm_layer(int(ngf * mult / 2)),
                      nn.ReLU(True)]
        model += [nn.ReflectionPad2d(3)]
        model += [nn.Conv2d(ngf, output_nc, kernel_size=7, padding=0)]
        model += [nn.Tanh()]

        self.model = nn.Sequential(*model)

    def forward(self, input):
        return self.model(input)


class UnetGenerator(nn.Module):
    """U-Net生成器"""
    def __init__(self, input_nc, output_nc, num_downs, ngf=64, 
                 norm_layer=nn.BatchNorm2d, use_dropout=False):
        super(UnetGenerator, self).__init__()
        unet_block = UnetSkipConnectionBlock(ngf * 8, ngf * 8, input_nc=None, 
                                            submodule=None, norm_layer=norm_layer, innermost=True)
        for i in range(num_downs - 5):
            unet_block = UnetSkipConnectionBlock(ngf * 8, ngf * 8, input_nc=None, 
                                                submodule=unet_block, norm_layer=norm_layer, use_dropout=use_dropout)
        unet_block = UnetSkipConnectionBlock(ngf * 4, ngf * 8, input_nc=None, submodule=unet_block, norm_layer=norm_layer)
        unet_block = UnetSkipConnectionBlock(ngf * 2, ngf * 4, input_nc=None, submodule=unet_block, norm_layer=norm_layer)
        unet_block = UnetSkipConnectionBlock(ngf, ngf * 2, input_nc=None, submodule=unet_block, norm_layer=norm_layer)
        self.model = UnetSkipConnectionBlock(output_nc, ngf, input_nc=input_nc, 
                                           submodule=unet_block, outermost=True, norm_layer=norm_layer)

    def forward(self, input):
        return self.model(input)


class UnetSkipConnectionBlock(nn.Module):
    """U-Net子模块"""
    def __init__(self, outer_nc, inner_nc, input_nc=None,
                 submodule=None, outermost=False, innermost=False, 
                 norm_layer=nn.BatchNorm2d, use_dropout=False):
        super(UnetSkipConnectionBlock, self).__init__()
        self.outermost = outermost
        if type(norm_layer) == functools.partial:
            use_bias = norm_layer.func == nn.InstanceNorm2d
        else:
            use_bias = norm_layer == nn.InstanceNorm2d
            
        if input_nc is None:
            input_nc = outer_nc
            
        downconv = nn.Conv2d(input_nc, inner_nc, kernel_size=4,
                             stride=2, padding=1, bias=use_bias)
        downrelu = nn.LeakyReLU(0.2, True)
        downnorm = norm_layer(inner_nc)
        uprelu = nn.ReLU(True)
        upnorm = norm_layer(outer_nc)

        if outermost:
            upconv = nn.ConvTranspose2d(inner_nc * 2, outer_nc,
                                        kernel_size=4, stride=2,
                                        padding=1)
            down = [downconv]
            up = [uprelu, upconv, nn.Tanh()]
            model = down + [submodule] + up
        elif innermost:
            upconv = nn.ConvTranspose2d(inner_nc, outer_nc,
                                        kernel_size=4, stride=2,
                                        padding=1, bias=use_bias)
            down = [downrelu, downconv]
            up = [uprelu, upconv, upnorm]
            model = down + up
        else:
            upconv = nn.ConvTranspose2d(inner_nc * 2, outer_nc,
                                        kernel_size=4, stride=2,
                                        padding=1, bias=use_bias)
            down = [downrelu, downconv, downnorm]
            up = [uprelu, upconv, upnorm]

            if use_dropout:
                model = down + [submodule] + up + [nn.Dropout(0.5)]
            else:
                model = down + [submodule] + up

        self.model = nn.Sequential(*model)

    def forward(self, x):
        if self.outermost:
            return self.model(x)
        else:
            return torch.cat([x, self.model(x)], 1)


class NLayerDiscriminator(nn.Module):
    """PatchGAN判别器"""
    def __init__(self, input_nc, ndf=64, n_layers=3, norm_layer=nn.BatchNorm2d):
        super(NLayerDiscriminator, self).__init__()
        if type(norm_layer) == functools.partial:
            use_bias = norm_layer.func == nn.InstanceNorm2d
        else:
            use_bias = norm_layer == nn.InstanceNorm2d

        kw = 4
        padw = 1
        sequence = [
            nn.Conv2d(input_nc, ndf, kernel_size=kw, stride=2, padding=padw),
            nn.LeakyReLU(0.2, True)
        ]

        nf_mult = 1
        nf_mult_prev = 1
        for n in range(1, n_layers):
            nf_mult_prev = nf_mult
            nf_mult = min(2 ** n, 8)
            sequence += [
                nn.Conv2d(ndf * nf_mult_prev, ndf * nf_mult,
                          kernel_size=kw, stride=2, padding=padw, bias=use_bias),
                norm_layer(ndf * nf_mult),
                nn.LeakyReLU(0.2, True)
            ]

        nf_mult_prev = nf_mult
        nf_mult = min(2 ** n_layers, 8)
        sequence += [
            nn.Conv2d(ndf * nf_mult_prev, ndf * nf_mult,
                      kernel_size=kw, stride=1, padding=padw, bias=use_bias),
            norm_layer(ndf * nf_mult),
            nn.LeakyReLU(0.2, True)
        ]

        sequence += [nn.Conv2d(ndf * nf_mult, 1, kernel_size=kw, stride=1, padding=padw)]
        self.model = nn.Sequential(*sequence)

    def forward(self, input):
        return self.model(input)


class Pix2PixModel(nn.Module):
    """Pix2Pix完整模型"""
    def __init__(self, input_nc=3, output_nc=3, ngf=64, ndf=64, 
                 netG='unet_256', netD='basic', n_layers_D=3, 
                 norm='batch', use_dropout=False, init_type='normal', 
                 init_gain=0.02, gan_mode='lsgan', lr=0.0002, beta1=0.5, 
                 lambda_L1=100.0, device='cuda'):
        super(Pix2PixModel, self).__init__()
        self.device = device
        
        # 定义生成器
        norm_layer = self.get_norm_layer(norm_type=norm)
        
        if netG == 'resnet_9blocks':
            self.netG = ResnetGenerator(input_nc, output_nc, ngf, norm_layer=norm_layer, 
                                       use_dropout=use_dropout, n_blocks=9)
        elif netG == 'resnet_6blocks':
            self.netG = ResnetGenerator(input_nc, output_nc, ngf, norm_layer=norm_layer, 
                                       use_dropout=use_dropout, n_blocks=6)
        elif netG == 'unet_128':
            self.netG = UnetGenerator(input_nc, output_nc, 7, ngf, norm_layer=norm_layer, 
                                     use_dropout=use_dropout)
        elif netG == 'unet_256':
            self.netG = UnetGenerator(input_nc, output_nc, 8, ngf, norm_layer=norm_layer, 
                                     use_dropout=use_dropout)
        else:
            raise NotImplementedError('Generator model name [%s] is not recognized' % netG)
        
        self.netG = self.init_net(self.netG, init_type, init_gain)
        
        # 定义判别器
        if netD == 'basic':
            self.netD = NLayerDiscriminator(input_nc + output_nc, ndf, n_layers=3, norm_layer=norm_layer)
        elif netD == 'n_layers':
            self.netD = NLayerDiscriminator(input_nc + output_nc, ndf, n_layers_D, norm_layer=norm_layer)
        else:
            raise NotImplementedError('Discriminator model name [%s] is not recognized' % netD)
        
        self.netD = self.init_net(self.netD, init_type, init_gain)
        
        # 定义损失函数
        self.criterionGAN = GANLoss(gan_mode).to(device)
        self.criterionL1 = nn.L1Loss()
        self.lambda_L1 = lambda_L1
        
        # 定义优化器
        self.optimizer_G = optim.Adam(self.netG.parameters(), lr=lr, betas=(beta1, 0.999))
        self.optimizer_D = optim.Adam(self.netD.parameters(), lr=lr, betas=(beta1, 0.999))
        
        # 损失记录
        self.loss_G = 0
        self.loss_D = 0
        self.loss_G_GAN = 0
        self.loss_G_L1 = 0
        self.loss_D_real = 0
        self.loss_D_fake = 0
        
    def get_norm_layer(self, norm_type='instance'):
        """返回归一化层"""
        if norm_type == 'batch':
            norm_layer = functools.partial(nn.BatchNorm2d, affine=True, track_running_stats=True)
        elif norm_type == 'instance':
            norm_layer = functools.partial(nn.InstanceNorm2d, affine=False, track_running_stats=False)
        elif norm_type == 'none':
            norm_layer = lambda x: nn.Identity()
        else:
            raise NotImplementedError('normalization layer [%s] is not found' % norm_type)
        return norm_layer
    
    def init_net(self, net, init_type='normal', init_gain=0.02):
        """初始化网络权重"""
        def init_func(m):
            classname = m.__class__.__name__
            if hasattr(m, 'weight') and (classname.find('Conv') != -1 or classname.find('Linear') != -1):
                if init_type == 'normal':
                    init.normal_(m.weight.data, 0.0, init_gain)
                elif init_type == 'xavier':
                    init.xavier_normal_(m.weight.data, gain=init_gain)
                elif init_type == 'kaiming':
                    init.kaiming_normal_(m.weight.data, a=0, mode='fan_in')
                elif init_type == 'orthogonal':
                    init.orthogonal_(m.weight.data, gain=init_gain)
                if hasattr(m, 'bias') and m.bias is not None:
                    init.constant_(m.bias.data, 0.0)
            elif classname.find('BatchNorm2d') != -1:
                init.normal_(m.weight.data, 1.0, init_gain)
                init.constant_(m.bias.data, 0.0)
        
        net.apply(init_func)
        return net
    
    def set_input(self, input):
        """设置输入数据"""
        self.real_A = input['A'].to(self.device)
        self.real_B = input['B'].to(self.device)
    
    def forward(self):
        """前向传播"""
        self.fake_B = self.netG(self.real_A)
    
    def backward_D(self):
        """判别器反向传播"""
        # 真实图像
        real_AB = torch.cat((self.real_A, self.real_B), 1)
        pred_real = self.netD(real_AB)
        self.loss_D_real = self.criterionGAN(pred_real, True)
        
        # 生成图像
        fake_AB = torch.cat((self.real_A, self.fake_B.detach()), 1)
        pred_fake = self.netD(fake_AB)
        self.loss_D_fake = self.criterionGAN(pred_fake, False)
        
        # 组合损失
        self.loss_D = (self.loss_D_real + self.loss_D_fake) * 0.5
        self.loss_D.backward()
    
    def backward_G(self):
        """生成器反向传播"""
        # GAN损失
        fake_AB = torch.cat((self.real_A, self.fake_B), 1)
        pred_fake = self.netD(fake_AB)
        self.loss_G_GAN = self.criterionGAN(pred_fake, True)
        
        # L1损失
        self.loss_G_L1 = self.criterionL1(self.fake_B, self.real_B) * self.lambda_L1
        
        # 组合损失
        self.loss_G = self.loss_G_GAN + self.loss_G_L1
        self.loss_G.backward()
    
    def optimize_parameters(self):
        """优化一步"""
        # 前向传播
        self.forward()
        
        # 更新判别器
        self.set_requires_grad(self.netD, True)
        self.optimizer_D.zero_grad()
        self.backward_D()
        self.optimizer_D.step()
        
        # 更新生成器
        self.set_requires_grad(self.netD, False)
        self.optimizer_G.zero_grad()
        self.backward_G()
        self.optimizer_G.step()
    
    def set_requires_grad(self, nets, requires_grad=False):
        """设置网络是否需要梯度"""
        if not isinstance(nets, list):
            nets = [nets]
        for net in nets:
            if net is not None:
                for param in net.parameters():
                    param.requires_grad = requires_grad
                    
    def get_current_losses(self):
        """返回当前损失"""
        return {
            'G': self.loss_G.item(),
            'D': self.loss_D.item(),
            'G_GAN': self.loss_G_GAN.item(),
            'G_L1': self.loss_G_L1.item(),
            'D_real': self.loss_D_real.item(),
            'D_fake': self.loss_D_fake.item()
        }
    
    def get_current_visuals(self):
        """返回当前可视化结果"""
        return {
            'real_A': self.real_A,
            'fake_B': self.fake_B,
            'real_B': self.real_B
        }

# -------------------------- 数据集处理（保持不变） --------------------------
class AlignedDataset(Dataset):
    """对齐数据集（修复版本）"""
    def __init__(self, root, phase='train', load_size=286, crop_size=256, 
                 preprocess='resize_and_crop', no_flip=False, input_nc=3,
                 val_split=0.1):
        super().__init__()
        self.root = root
        self.phase = phase
        self.dir_AB = os.path.join(root, phase)
        self.load_size = load_size
        self.crop_size = crop_size
        self.preprocess = preprocess
        self.no_flip = no_flip
        self.input_nc = input_nc
        self.val_split = val_split
        
        # 先检查当前phase目录是否存在
        if not os.path.exists(self.dir_AB):
            print(f"警告：{self.dir_AB} 目录不存在，将从train目录划分{val_split*100}%作为{phase}集")
            self.dir_AB = os.path.join(root, 'train')
            all_paths = sorted(self.make_dataset(self.dir_AB))
            if len(all_paths) == 0:
                raise ValueError("错误：train目录也无有效数据")
            
            split_idx = int(len(all_paths) * (1 - val_split))
            self.AB_paths = all_paths[split_idx:] if phase == 'val' else all_paths[:split_idx]
        else:
            # 目录存在，尝试加载
            self.AB_paths = sorted(self.make_dataset(self.dir_AB))
            # 若目录存在但无有效文件，仍从train划分
            if len(self.AB_paths) == 0:
                print(f"警告：{self.dir_AB} 目录无有效文件，从train划分{val_split*100}%作为{phase}集")
                self.dir_AB = os.path.join(root, 'train')
                all_paths = sorted(self.make_dataset(self.dir_AB))
                split_idx = int(len(all_paths) * (1 - val_split))
                self.AB_paths = all_paths[split_idx:] if phase == 'val' else all_paths[:split_idx]
        
        self.transform = self.get_transform()
        if len(self.AB_paths) == 0:
            raise ValueError(f"错误：无法获取有效的{phase}集数据")
        print(f"成功加载 {len(self.AB_paths)} 个有效图像文件（{phase}集）")
    
    def is_image_file(self, filename):
        """判断是否为图像文件"""
        IMG_EXTENSIONS = ['.jpg', '.JPG', '.jpeg', '.JPEG', '.png', '.PNG', '.bmp', '.BMP']
        return any(filename.endswith(ext) for ext in IMG_EXTENSIONS)
    
    def make_dataset(self, dir):
        """创建数据集路径列表（修复目录检查）"""
        images = []
        if not os.path.isdir(dir):
            return images  # 目录不存在时返回空列表
        
        for root, _, fnames in sorted(os.walk(dir)):
            for fname in fnames:
                if self.is_image_file(fname):
                    path = os.path.join(root, fname)
                    if os.path.getsize(path) == 0:
                        continue
                    try:
                        with Image.open(path) as img:
                            img.convert('RGB')
                        images.append(path)
                    except:
                        continue
        return images
    
    def get_transform(self):
        """获取图像转换管道"""
        transform_list = []
        if 'resize' in self.preprocess:
            transform_list.append(transforms.Resize((self.load_size, self.load_size), Image.BICUBIC))
        if 'crop' in self.preprocess:
            transform_list.append(transforms.RandomCrop(self.crop_size))
        if not self.no_flip:
            transform_list.append(transforms.RandomHorizontalFlip())
        transform_list += [transforms.ToTensor()]
        if self.input_nc == 1:
            transform_list += [transforms.Normalize((0.5,), (0.5,))]
        else:
            transform_list += [transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
        return transforms.Compose(transform_list)
    
    def __len__(self):
        """返回数据集大小"""
        return len(self.AB_paths)
    
    def __getitem__(self, index):
        """获取样本"""
        while True:
            try:
                AB_path = self.AB_paths[index]
                with Image.open(AB_path) as img:
                    AB = img.convert('RGB')
                w, h = AB.size
                w2 = int(w / 2)
                A = AB.crop((w2, 0, w, h))  # 右侧：输入
                B = AB.crop((0, 0, w2, h))  # 左侧：目标
                if self.input_nc == 1:
                    A = A.convert('L')
                A = self.transform(A)
                B = self.transform(B)
                return {'A': A, 'B': B, 'path': AB_path}
            except:
                index = (index + 1) % len(self.AB_paths)

# -------------------------- 多任务配置（保持不变） --------------------------
TASK_CONFIGS = {
    'maps': {
        'name': 'Aerial to Map',
        'input_nc': 3,
        'output_nc': 3,
        'dataset_dir': 'datasets/datasets/maps',
        'netG': 'unet_256',
        'epochs': 100
    },
    'night2day': {
        'name': 'Day to Night',
        'input_nc': 3,
        'output_nc': 3,
        'dataset_dir': 'datasets/datasets/night2day',
        'netG': 'unet_256',
        'epochs': 150,
        'direction': 'BtoA'
    },
    'edges2photos': {
        'name': 'edges to photos',
        'input_nc': 1,
        'output_nc': 3,
        'dataset_dir': 'datasets/datasets/edges2photos',
        'netG': 'unet_256',
        'epochs': 120
    },
    'facades': {
        'name': 'labels to facade',
        'input_nc': 3,
        'output_nc': 3,
        'dataset_dir': 'datasets/datasets/facades',
        'netG': 'unet_256',
        'epochs': 100
    }
}

# -------------------------- 训练与测试函数（简化进度条） --------------------------
def train_task(task, save_dir='checkpoints'):
    """训练指定任务的模型（简化进度条）"""
    cfg = TASK_CONFIGS[task]
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"使用设备: {device}")
    
    # 1. 加载数据集
    train_dataset = AlignedDataset(
        root=cfg['dataset_dir'],
        phase='train',
        input_nc=cfg['input_nc']
    )
    val_dataset = AlignedDataset(
        root=cfg['dataset_dir'],
        phase='val',
        input_nc=cfg['input_nc']
    )
    print(f"训练集大小: {len(train_dataset)}, 验证集大小: {len(val_dataset)}")
    
    # 2. 初始化模型
    model = Pix2PixModel(
        input_nc=cfg['input_nc'],
        output_nc=cfg['output_nc'],
        netG=cfg['netG'],
        device=device
    )
    model.to(device)
    
    # 3. 训练模型
    save_path = os.path.join(save_dir, task)
    os.makedirs(save_path, exist_ok=True)
    dataloader = DataLoader(train_dataset, batch_size=1, shuffle=True, num_workers=0)
    total_batches = len(dataloader)
    
    print(f"开始训练 {cfg['name']} 任务...")
    # 仅保留单轮次进度条
    for epoch in tqdm(range(cfg['epochs']), desc=f"{cfg['name']} 训练进度"):
        epoch_start = time.time()
        total_loss_G = 0.0
        total_loss_D = 0.0
        
        # 移除批次进度条，仅在轮次内循环
        for i, data in enumerate(dataloader):
            model.set_input(data)
            model.optimize_parameters()
            
            # 累计损失
            losses = model.get_current_losses()
            total_loss_G += losses['G']
            total_loss_D += losses['D']
        
        # 计算平均损失
        avg_loss_G = total_loss_G / total_batches
        avg_loss_D = total_loss_D / total_batches
        
        # 每10轮打印一次详细信息（减少输出频率）
        if (epoch + 1) % 10 == 0 or epoch == 0:
            print(f"Epoch {epoch+1}/{cfg['epochs']} | G_avg: {avg_loss_G:.4f} | D_avg: {avg_loss_D:.4f} | 耗时: {time.time()-epoch_start:.1f}s")
        
        # 保存模型
        if (epoch + 1) % 20 == 0:
            torch.save({
                'epoch': epoch + 1,
                'netG_state_dict': model.netG.state_dict(),
                'netD_state_dict': model.netD.state_dict(),
                'losses': {'G': avg_loss_G, 'D': avg_loss_D}
            }, os.path.join(save_path, f"epoch_{epoch+1}.pth"))
    
    print(f"{cfg['name']} 任务训练完成，模型保存至 {save_path}")
    return model


def test_task(task, model=None, num_test=10):
    """测试指定任务的模型（简化输出）"""
    cfg = TASK_CONFIGS[task]
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    # 加载数据集
    test_dataset = AlignedDataset(
        root=cfg['dataset_dir'],
        phase='val',
        input_nc=cfg['input_nc']
    )
    dataloader = DataLoader(test_dataset, batch_size=1, shuffle=False, num_workers=0)
    
    # 加载模型
    if model is None:
        model = Pix2PixModel(
            input_nc=cfg['input_nc'],
            output_nc=cfg['output_nc'],
            netG=cfg['netG'],
            device=device
        )
        model.to(device)
        checkpoint_dir = os.path.join('checkpoints', task)
        if os.path.exists(checkpoint_dir) and len(os.listdir(checkpoint_dir)) > 0:
            latest_ckpt = sorted(os.listdir(checkpoint_dir))[-1]
            ckpt = torch.load(os.path.join(checkpoint_dir, latest_ckpt), map_location=device)
            model.netG.load_state_dict(ckpt['netG_state_dict'])
        else:
            print(f"警告：未找到 {task} 的模型 checkpoint，使用随机初始化模型测试")
    
    # 测试（简化进度显示）
    results = []
    model.eval()
    with torch.no_grad():
        for i in tqdm(range(min(num_test, len(dataloader))), desc=f"{cfg['name']} 测试进度"):
            data = next(iter(dataloader))
            model.set_input(data)
            model.forward()
            results.append(model.get_current_visuals())
    
    return results


def visualize_task_results(results, task):
    """可视化任务测试结果（保持不变）"""
    cfg = TASK_CONFIGS[task]
    plt.figure(figsize=(15, 5 * len(results)))
    
    for i, visuals in enumerate(results):
        real_A = tensor2im(visuals['real_A'])  # 输入
        fake_B = tensor2im(visuals['fake_B'])  # 输出
        real_B = tensor2im(visuals['real_B'])  # 真实目标
        
        plt.subplot(len(results), 3, i*3 + 1)
        plt.title("输入")
        plt.imshow(real_A)
        plt.axis('off')
        
        plt.subplot(len(results), 3, i*3 + 2)
        plt.title("生成结果")
        plt.imshow(fake_B)
        plt.axis('off')
        
        plt.subplot(len(results), 3, i*3 + 3)
        plt.title("真实目标")
        plt.imshow(real_B)
        plt.axis('off')
    
    plt.suptitle(f"{cfg['name']} 任务结果对比", y=1.02)
    plt.tight_layout()
    plt.show()


# -------------------------- 工具函数（保持不变） --------------------------
def tensor2im(input_image, imtype=np.uint8):
    """将Tensor转换为图像"""
    if isinstance(input_image, torch.Tensor):
        image_tensor = input_image.data
        image_numpy = image_tensor[0].cpu().float().numpy()
        if image_numpy.shape[0] == 1:  # 单通道转3通道
            image_numpy = np.tile(image_numpy, (3, 1, 1))
        image_numpy = (np.transpose(image_numpy, (1, 2, 0)) + 1) / 2.0 * 255.0
    return image_numpy.astype(imtype)


# -------------------------- 主函数（保持不变） --------------------------
def main():
    # 获取所有任务
    tasks = list(TASK_CONFIGS.keys())
    
    # 按顺序训练每个任务
    for task in tasks:
        cfg = TASK_CONFIGS[task]
        print(f"\n{'='*50}\n开始训练任务: {cfg['name']} (数据集: {cfg['dataset_dir']})\n{'='*50}")
        
        # 检查数据集是否存在
        if not os.path.exists(cfg['dataset_dir']):
            print(f"错误：未找到数据集 {cfg['dataset_dir']}，跳过此任务")
            continue
        
        # 训练模型
        try:
            model = train_task(task)
            print(f"{cfg['name']} 模型训练完成")
            
            # 测试并可视化
            test_enabled = True  # 可设置为False跳过测试
            if test_enabled:
                print(f"测试 {cfg['name']} 模型...")
                results = test_task(task, model, num_test=3)
                visualize_task_results(results, task)
        except Exception as e:
            print(f"{cfg['name']} 任务训练出错：{e}")
            print(f"跳过此任务，继续下一个...")
            continue
    
    print("\n所有任务训练完成！")


if __name__ == "__main__":
    main()
```
